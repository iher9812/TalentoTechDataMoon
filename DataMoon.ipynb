## 1. Calcular el gasto anual en los productos de Gestión menstrual:

```python
import pandas as pd

# Update the file path to point to the downloaded Excel file
file_path = '/content/Base de datos productos de gestion mestrual.xlsx'
df_dict = pd.read_excel(file_path, sheet_name=None)
```

```python
# Display the first few rows of the dataframe
df = df_dict['Toallas']
df.head()
```

### Toallas y Tampones

```python
# Load the sheets
toallas_df = pd.read_excel(file_path, sheet_name='Toallas')
tampones_df = pd.read_excel(file_path, sheet_name='Tampones')

# Display the first few rows of each dataframe to understand the structure
toallas_df.head(), tampones_df.head()
```

```python
# Calcular la cantidad de productos usados anualmente
toallas_por_dia = 4
dias_por_mes_toallas = 5
tampones_por_dia = 6
dias_por_mes_tampones = 5
meses_por_ano = 12

toallas_uso_anual = toallas_por_dia * dias_por_mes_toallas * meses_por_ano
tampones_uso_anual = tampones_por_dia * dias_por_mes_tampones * meses_por_ano

# Print the results
print("Toallas per year:", toallas_uso_anual)
print("Tampones per year:", tampones_uso_anual)
```

```python
# Calcular el costo promedio anual de los productos en cada hoja
toallas_df['Promedio'] = toallas_df[['Nosotras', 'Kotex', 'Fresh&Free']].mean(axis=1)
tampones_df['Promedio'] = tampones_df[['Kotex digital super', 'O.B. Original Medio', 'Nosotras Super']].mean(axis=1)

# Calcular el costo total anual para cada año
toallas_df['Costo Anual'] = toallas_df['Promedio'] * toallas_uso_anual
tampones_df['Costo Anual'] = tampones_df['Promedio'] * tampones_uso_anual

# Mostrar los resultados
toallas_df[['Año', 'Promedio', 'Costo Anual']], tampones_df[['Año', 'Promedio', 'Costo Anual']]
```

```python
import matplotlib.pyplot as plt

# Graficar los resultados para las toallas
plt.figure(figsize=(10, 6))
plt.plot(toallas_df['Año'], toallas_df['Costo Anual'], marker='o', linestyle='-', color='b')
plt.title('Costo Anual Promedio de Toallas a lo largo de los Años')
plt.xlabel('Año')
plt.ylabel('Costo Anual (COP)')
plt.grid(True)
plt.xticks(toallas_df['Año'])
plt.tight_layout()
plt.show()
```

```python
# Graficar los resultados para los tampones
plt.figure(figsize=(10, 6))
plt.plot(tampones_df['Año'], tampones_df['Costo Anual'], marker='o', linestyle='-', color='r')
plt.title('Costo Anual Promedio de Tampones a lo largo de los Años')
plt.xlabel('Año')
plt.ylabel('Costo Anual (COP)')
plt.grid(True)
plt.xticks(tampones_df['Año'])
plt.tight_layout()
plt.show()
```

### Copas mestruales

```python
# Update the file path to point to the downloaded Excel file
file_path = '/content/Base de datos productos de gestion mestrual.xlsx'
copas_df= pd.read_excel(file_path, sheet_name='COPAS')

# Mostrar las primeras filas del dataframe para verificar su contenido
copas_df.head()


```

```python
# Calcular el número de copas menstruales necesarias por año
# Si una copa dura 5 años, entonces se necesita 1/5 copas por año
copas_df['Copas_por_año'] = 1
```

```python
# Calcular el costo promedio anual de las copas menstruales
# Promediar los precios de las columnas UVA, LifeCup y Nosotras
copas_df['Costo_promedio'] = copas_df[['UVA', 'LifeCup', 'Nosotras Cup']].mean(axis=1)
```

```python
# Mostrar los resultados
copas_df[['Año', 'Copas_por_año', 'Costo_promedio']]
```

```python
#Numero promedio de copas durante el periodo de 2016 al 2024

import math
# Número total de años en la tabla
total_years = copas_df.shape[0]

# Duración de una copa menstrual en años
copa_duration = 5 #años

# Calcular el número de copas necesarias
copas_necesarias = total_years / copa_duration

# Redondear hacia arriba para obtener un número entero de copas

copas_necesarias = math.ceil(copas_necesarias)

print(f'Necesitarás aproximadamente {copas_necesarias} copas menstruales para un periodo de {total_years} años.')
```

```python
# Calcular el gasto anual en copas menstruales
copas_df['gasto_total'] = copas_necesarias * copas_df['Costo_promedio'] # Use the variable copas_necesarias directly
print(f'El gasto total en copas menstruales desde 2016 hasta 2024 es: ${copas_df["gasto_total"].mean()}') # Calculate and print the total cost
```

```python
# Crear una gráfica que muestre el Costo_promedio por año
plt.figure(figsize=(10, 6))
plt.plot(copas_df['Año'], copas_df['Costo_promedio'], marker='o', linestyle='-', color='b')
plt.title('Costo Anual Promedio de las Copas a lo largo de los Años')
plt.xlabel('Año')
plt.ylabel('Costo Anual (COP)')
plt.grid(True)
plt.show()
```

## 3. Calcular el Ahorro Anual por la Eliminación del IVA:

```python
import matplotlib.pyplot as plt
# Crear un DataFrame para los costos anuales originales
precios_originales = pd.DataFrame({
    'Año': toallas_df['Año'],
    'Toallas': toallas_df['Costo Anual'],
    'Tampones': tampones_df['Costo Anual'],
    'Copas': copas_df['Costo_promedio']
})
```

```python
# Aplicar un aumento del 19% a todos los productos hasta 2018
precios_con_aumento = precios_originales.copy()
precios_con_aumento[['Toallas', 'Tampones', 'Copas']] *= 1.19
```

```python
# Crear la tabla comparativa usando diferencia en diferencia
did_df = pd.DataFrame({
    'Año': precios_originales['Año'],
    'Toallas_original': precios_originales['Toallas'],
    'Toallas_con_aumento': precios_con_aumento['Toallas'],
    'Tampones_original': precios_originales['Tampones'],
    'Tampones_con_aumento': precios_con_aumento['Tampones'],
    'Copas_original': precios_originales['Copas'],
    'Copas_con_aumento': precios_con_aumento['Copas'],
})
```

```python
# Calcular cómo se vería la eliminación del 19% a partir de 2018
did_df['Toallas_DiD'] = did_df['Toallas_con_aumento']
did_df['Tampones_DiD'] = did_df['Tampones_con_aumento']
did_df['Copas_DiD'] = did_df['Copas_con_aumento']
```

```python
# Aplicar la eliminación del aumento solo en 2018 y 2019
did_df.loc[did_df['Año'] >= 2019, 'Toallas_DiD'] = did_df['Toallas_con_aumento'] / 1.19
did_df.loc[did_df['Año'] >= 2019, 'Tampones_DiD'] = did_df['Tampones_con_aumento'] / 1.19
did_df.loc[did_df['Año'] >= 2019, 'Copas_DiD'] = did_df['Copas_con_aumento'] / 1.19
```

```python
# Mostrar la tabla comparativa
print("Tabla comparativa de precios con y sin aumento utilizando diferencia en diferencia:")
print(did_df)
```

```python
# Graficar resultados
plt.figure(figsize=(12, 6))
plt.plot(did_df['Año'], did_df['Toallas_original'], marker='o', linestyle='-', color='blue', label='Toallas original')
plt.plot(did_df['Año'], did_df['Toallas_con_aumento'], marker='o', linestyle='--', color='cyan', label='Toallas con aumento')
plt.plot(did_df['Año'], did_df['Toallas_DiD'], marker='o', linestyle=':', color='navy', label='Toallas DiD (a partir del 2018)')
plt.title('Comparación de Costo Anual de Toallas con y sin Aumento (DiD)')
plt.xlabel('Año')
plt.ylabel('Costo Anual (COP)')
plt.grid(True)
plt.legend()
plt.xticks(did_df['Año'])
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(did_df['Año'], did_df['Tampones_original'], marker='o', linestyle='-', color='red', label='Tampones original')
plt.plot(did_df['Año'], did_df['Tampones_con_aumento'], marker='o', linestyle='--', color='magenta', label='Tampones con aumento')
plt.plot(did_df['Año'], did_df['Tampones_DiD'], marker='o', linestyle=':', color='darkred', label='Tampones DiD (a partir del 2018)')
plt.title('Comparación de Costo Anual de Tampones con y sin Aumento (DiD)')
plt.xlabel('Año')
plt.ylabel('Costo Anual (COP)')
plt.grid(True)
plt.legend()
plt.xticks(did_df['Año'])
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(did_df['Año'], did_df['Copas_original'], marker='o', linestyle='-', color='green', label='Copas original')
plt.plot(did_df['Año'], did_df['Copas_con_aumento'], marker='o', linestyle='--', color='lime', label='Copas con aumento')
plt.plot(did_df['Año'], did_df['Copas_DiD'], marker='o', linestyle=':', color='darkgreen', label='Copas DiD (a partir del 2018)')
plt.title('Comparación de Costo Anual de Copas con y sin Aumento (DiD)')
plt.xlabel('Año')
plt.ylabel('Costo Anual (COP)')
plt.grid(True)
plt.legend()
plt.xticks(did_df['Año'])
plt.tight_layout()
plt.show()
```

##  Análisis: Estimar el ahorro anual total y por persona menstruante que se obtendría al eliminar el IVA de los productos de gestión menstrual entre 2016 y 2021

### Toallas y tampones

```python
# Cargar el archivo Excel
file_path = '/content/precios_comparacion_did.xlsx'
df = pd.read_excel(file_path)
```

```python
# Calcular los totales anuales para cada columna, excluyendo copas
totales_toallas = df[['Toallas_original', 'Toallas_con_aumento']].sum()
totales_tampones = df[['Tampones_original', 'Tampones_con_aumento']].sum()
```

```python
# Crear un DataFrame para los totales
totales_df = pd.DataFrame({
    'Concepto': ['Total Toallas_original', 'Total Toallas_con_aumento',
                 'Total Tampones_original', 'Total Tampones_con_aumento'],
    'Valor': [totales_toallas['Toallas_original'], totales_toallas['Toallas_con_aumento'],
              totales_tampones['Tampones_original'], totales_tampones['Tampones_con_aumento']]
})
```

```python
# Calcular la diferencia entre los totales originales y con aumento
diferencias = {
    'Concepto': ['Diferencia Toallas', 'Diferencia Tampones'],
    'Valor': [
        totales_toallas['Toallas_con_aumento'] - totales_toallas['Toallas_original'],
        totales_tampones['Tampones_con_aumento'] - totales_tampones['Tampones_original']
    ]
}

diferencias_df = pd.DataFrame(diferencias)
```

```python
# Mostrar los resultados
print("Totales Anuales:")
print(totales_df)
print("\nDiferencias entre Original y con Aumento:")
print(diferencias_df)
```

```python
# Extraer los valores de los totales calculados
total_toallas_original = totales_df.loc[totales_df['Concepto'] == 'Total Toallas_original', 'Valor'].values[0]
total_tampones_original = totales_df.loc[totales_df['Concepto'] == 'Total Tampones_original', 'Valor'].values[0]

# Imprimir el mensaje
print(f"Durante el periodo del 2016 al 2024 (9 años) se gastó un promedio de {total_toallas_original:.2f} en toallas y un promedio de {total_tampones_original:.2f} en tampones. ")
```

### Copas

```python
# Calcular los totales anuales para Copas
totales_copas = df[['Copas_original', 'Copas_con_aumento']].sum()
```

```python
# Calcular el promedio de Copas_original y Copas_con_aumento
promedio_copas_original = totales_copas['Copas_original'] / len(df)
promedio_copas_con_aumento = totales_copas['Copas_con_aumento'] / len(df)
```

```python
# Multiplicar los promedios por dos
promedio_copas_original_doble = promedio_copas_original * 2
promedio_copas_con_aumento_doble = promedio_copas_con_aumento * 2
```

```python
# Restar los resultados
resultado = promedio_copas_con_aumento_doble - promedio_copas_original_doble
```

```python
# Imprimir el resultado
print(f"Durante el periodo del 2016 al 2024 (9 años) se gastó un promedio de {promedio_copas_original_doble:.2f} en copas mestruales. \nCon el aumento del IVA se hubiera inclementado el precio y se habria gastado {promedio_copas_con_aumento_doble:.2f} un aunmento del valor de: {resultado:.2f} ")
```
